##**Function to get the article** 
This function uses to get the article from its URL.
Madina's news articles are under the  tag <'p'> always .

```python
def madinaNews(url):
  from urllib.request import urlopen
  html = urlopen(url).read().decode("utf8","ignore")#using utf8 because it is an arbic text
  soup =BeautifulSoup(html,"lxml")
  soup
  t= " ".join(map(lambda p: p.text, soup.find_all("p")))#the articles in madina news paper are under p tag
  return t
```

##**Summarize function**
this function wiil make series of steps :
1. making a space between the (.) and the next sentence so the sentence tokinize could work
2. break the text into lists of words
3. create a set of stop word to filter them from the text
4. create a frequency table of the words.
 
  A Python default dictionary is diffrent from regular dictionary such that if you search for a key that is not in the dictionart it will add it to the dictionary instead of throw error
5. significant score will assigned for each sentence in the article Depending on the words it contains and the frequency table.
6. finally , pick top n sentence based on the significant store then return the index of the important sentence

```python
def summarize(t):
  t= re.sub(r'(?<=[.,])(?=[^\s])', r' ', t)
  word_text=word_tokenize(t)
  arstop=set(stopwords.words("arabic")+list(punctuation))
  word_text=[w for w in word_text if w not in arstop]
  freq=FreqDist(word_text)
  nlargest(10,freq,key=freq.get)
  ranking=defaultdict(int)
  madina_sen=sent_tokenize(t)
  for i,sent in enumerate(madina_sen):
      for w in word_tokenize(sent.lower()):
        if w in freq: 
          ranking[i]+=freq[w]
  sen_index=nlargest(5,ranking,key=ranking.get)#return the index of the important sentence
  #the index is not in order so we have to print the sentence is acseding order
  return [madina_sen[i] for i in sorted(sen_index)]
```

```python
summ=summarize(t)
summ
```
